# ¿Influirá el incumplimiento en fechas de entrega en el score promedio?

Para esto se debe crear una columna que establezca un Sí o un No para el hecho de haber cumplido o no el tiempo de entrega. Se debe manejar el uso de fechas y operaciones de fechas en Pandas.

Revisamos los tipos de datos de todas las columnas de fecha, sea que vaya a trabajar con ellas ahora o no.

```python
df_orders.info()
```

Convertimos a datetime esas columnas...

```python
df_orders['order_estimated_delivery_date'] = pd.to_datetime(df_orders['order_estimated_delivery_date'])
df_orders['order_delivered_customer_date'] = pd.to_datetime(df_orders['order_delivered_customer_date'])
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])
df_orders['order_approved_at'] = pd.to_datetime(df_orders['order_approved_at'])
df_orders['order_delivered_carrier_date'] = pd.to_datetime(df_orders['order_delivered_carrier_date'])
df_orders.info()
```

Creamos la columna "delivered_on_time" que diga si se entregó a tiempo, **1 para Sí, 0 para No**

```python
df_orders['delivered_on_time'] = df_orders['order_estimated_delivery_date'] - df_orders['order_delivered_customer_date']
df_orders['delivered_on_time'] = df_orders['delivered_on_time'].apply(lambda x: 1 if x.days >= 0 else 0)
df_orders
```

Vamos a ver cuál es el porcentaje de entregas fallidas de la empresa:



```python
round(df_orders[df_orders['delivered_on_time'] == 0]['delivered_on_time'].count() / df_orders['delivered_on_time'].count() * 100, 1)
```

**De 100 envíos, 11 no se están entregando dentro de las fechas esperadas**. 

No es tan alto el porcentaje. A priori, pensamos que esto puede impactar negativamente en las calificaciones promedio de las categorías.

Ahora unimos los datasets involucrados para la consulta. Se me ocurre después separarlos según si se entrega a tiempo y si no.

```python
df_score_by_category = df_order_reviews.merge(df_orders, on='order_id').merge(df_order_items, on='order_id').merge(df_products, on='product_id')[['product_category_name', 'review_score', 'product_id', 'delivered_on_time']]

df_delivered_on_time = df_score_by_category[df_score_by_category['delivered_on_time'] == 1][['product_category_name', 'review_score', 'product_id']]
df_delivered_off_time = df_score_by_category[df_score_by_category['delivered_on_time'] == 0][['product_category_name', 'review_score', 'product_id']]

df_delivered_on_time = df_delivered_on_time.groupby('product_category_name')[['review_score', 'product_id']].agg({'review_score': np.mean, 'product_id': lambda x: x.count()}).sort_values('product_id', ascending=False)
a = df_delivered_on_time[df_delivered_on_time['product_id'] > 100].sort_values(by=['review_score'], ascending = False).reset_index().sort_values(by=['product_category_name'])#.head(10)

df_delivered_off_time = df_delivered_off_time.groupby('product_category_name')[['review_score', 'product_id']].agg({'review_score': np.mean, 'product_id': lambda x: x.count()}).sort_values('product_id', ascending=False)
b = df_delivered_off_time[df_delivered_off_time['product_id'] > 100].sort_values(by=['review_score'], ascending = False).reset_index().sort_values(by=['product_category_name'])#.head(10)
## Hacemos un merge de tipo inner para que no se tengan en cuenta las categorías que no hayan en común. 
# Esto quiere decir que hay categorías que o siempre han sido entregadas a tiempo o que siempre han sido entregadas a destiempo. 
# La salida nos mostrará que de 48 categorías, 5 no están en a y b al tiempo.
c_merge = a.merge(b, on='product_category_name', how='inner')
c_merge
```

Este hallazgo del Dataframe es interesante: **El promedio del score de una categoría, en general suele bajar si el producto es entregado fuera de tiempo**

Veamos esto de una manera gráfica:

```python
fig, axes = plt.subplots(figsize=(20,6))
axes.plot(c_merge['product_category_name'], c_merge['review_score_x'], 'r')
axes.plot(c_merge['product_category_name'], c_merge['review_score_y'], 'b')
plt.show()
```

```python
df_score_by_category = df_order_reviews.merge(df_orders, on='order_id').merge(df_order_items, on='order_id').merge(df_products, on='product_id')[['product_category_name', 'review_score', 'product_id', 'delivered_on_time']].groupby(['product_category_name', 'delivered_on_time'])[['review_score', 'product_id']].agg({'review_score': np.mean, 'product_id': lambda x: x.count()}).sort_values(by=['product_category_name'])
df_score_by_category = df_score_by_category.reset_index()
df_score_by_category = df_score_by_category[df_score_by_category['product_id'] >= 100]

plt.figure(figsize=(20,20))
sns.displot(data=df_score_by_category, x='product_category_name', y='review_score', hue='delivered_on_time', kind='hist')
plt.xticks(rotation=90)
plt.show()
```

# Ventas por mes

Necesitamos unir 2 datasets, uno que contiene el valor de la compra y otro que tiene la fecha.

```python
ventas_por_mes = df_orders.merge(df_order_payments, on='order_id', how='inner')
```

```python
ventas_por_mes['fecha_mes_anio'] = ventas_por_mes['order_purchase_timestamp'].dt.strftime(date_format="%Y-%m")
ventas_por_mes['fecha_mes_anio'] = pd.to_datetime(ventas_por_mes['fecha_mes_anio'])
ventas_por_mes = ventas_por_mes[['fecha_mes_anio', 'payment_value']]
ventas_por_mes = ventas_por_mes.groupby('fecha_mes_anio')[['payment_value']].sum().resample('M').sum().reset_index()
#ventas_por_mes['num_mes'] = ventas_por_mes['fecha_mes_anio'].
ventas_por_mes.head()
```

```python
ventas_por_mes.drop(range(24, 26), axis=0, inplace=True)
```

```python
ventas_por_mes.tail()
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(data = ventas_por_mes, x='fecha_mes_anio', y='payment_value')
plt.show()
```

Ahora vamos a emplear una regresión lineal

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
reg = LinearRegression()
# Se obtienen los arreglos para X e y
X = ventas_por_mes['fecha_mes_anio'].values.reshape(-1, 1)
y = ventas_por_mes['payment_value']

# Se estandarizan las variables
sc_x = StandardScaler()
sc_y = StandardScaler()
X = sc_x.fit_transform(X)
y = sc_y.fit_transform(y[:, np.newaxis]).flatten()

# Separamos el dataset en conjunto de preparación y conjunto de prueba
X_input, X_test, y_input, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Ajustamos el modelo con los datos de entrenamiento
model = reg.fit(X_input, y_input)
print('Intercepto: {:.3f}'.format(reg.intercept_))
print('Coeficientes:', reg.coef_)

# Calculamos la exactitud del modelo
print('Exactitud: {:.3f}'.format(reg.score(X_test, y_test)))

#predecimos con el 20%
y_grafica = model.predict(X_test)

# Guardamos a, y b para graficar con los puntos
a = round(reg.coef_[0], 3)
b = round(reg.intercept_, 3)
print(a,b)
```

```python
plt.plot(X_test, y_grafica, label='Regresión Lineal', color='blue')
plt.scatter(X_test, y_test, label='Datos de prueba', color='green')
plt.show()
```
